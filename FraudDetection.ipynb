```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from scipy.stats import norm
from scipy.stats import multivariate_normal
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
if df.isnull().values.any():
    print("Data has missing values")
```

```python
df.head()
```

```python
fig,ax=plt.subplots(1,len(df.columns),figsize=(len(df.columns)*5,5))
for i,column in enumerate(df.columns):
    #print(column)
    df[column].plot.hist(ax=ax[i],bins=100)
    ax[i].set_title(column)
plt.tight_layout()
```

```python
fig, axs = plt.subplots(6, 5, squeeze=False,figsize=(30,25))
for i, ax in enumerate(axs.flatten()):
    ax.set_title(df.columns[i])
    ax.set_xlabel(df.columns[i])
    sns.distplot(df.iloc[:, i], ax=ax, fit=norm,color="green", fit_kws={"color": "yellow"})
plt.tight_layout()
plt.show()
```
